```python
import pandas as pd
```

```python
df = pd.read_csv("tmdb-movies.csv")
```

```python
df.columns
```

```python
df_2006 = df[df['release_year']==2006]
```

```python
df_2006.isnull()
```

```python
df_2006.dropna(axis=0,how = 'any')
```

```python
df_2006_runtime = df_2006['runtime']
```

```python
df_2006_runtime
```

```python
df_2006_runtime.describe()
```

```python
sorted_df = df_2006_runtime.sort_values()
```

```python
sorted_df
```

```python
import numpy as np
```

```python
q1,q3 = np.percentile(sorted_df,[25,75])
```

```python
q1,q3
```

```python
iqr = q3-q1
iqr
```

```python
lower_bound = q1 - (1.5*iqr)
upper_bound = q3 + (1.5*iqr)
lower_bound,upper_bound
```

```python
sorted_df[(sorted_df>lower_bound)&(sorted_df<upper_bound)]
```

```python
sum = sorted_df.sum()
sum
```

```python
number_of_rows = sorted_df.shape
number_of_rows
```

```python
average = sum / number_of_rows
```
