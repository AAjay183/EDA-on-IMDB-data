```python
import pandas as pd
df = pd.read_csv("tmdb-movies.csv")
```

```python
df
```

```python
df.isnull()
```

```python
df1 = df.dropna(axis = 0, how = 'any')
```

```python
df1
```

```python
df2 = df1[(df1['release_year']>2000) & (df1['release_year']<2005)]
```

```python
df2
```

```python
df2['no_of_words'] = df2.original_title.apply(lambda x: len(x.split()))
```

```python
df2['no_of_words']
```

```python
total_no_of_words = df2['no_of_words'].sum()
```

```python
total_no_of_words
```
